```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
talep = [5,10,15,20,27,35,45,65,95,150,200]
fiyat = [5,10,15,20,25,30,35,40,45,50,55]
df = pd.DataFrame({"talep" : talep, "fiyat":fiyat})

plt.scatter(df.talep,
           df.fiyat,
           s=50,
           c="r")
plt.title("Talep ve Fiyat")
plt.xlabel("Talep")
plt.ylabel("Fiyat")
plt.grid(True)
plt.show();
```

```python
X = df[["talep"]]
y = df[["fiyat"]]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
model.intercept_
```

```python
plt.scatter(df.talep,
           df.fiyat,
           s=50,
           c="r")

plt.plot(X, model.predict(X))


plt.title("Talep ve Fiyat")
plt.xlabel("Talep")
plt.ylabel("Fiyat")
plt.grid(True)
plt.show();
```
